```python
def create_ref_points(eqns):
    reference_points = []
    var = ['x'] * (len(eqns[0]) - 1)
    for coeff in eqns:
      for i in range(len(coeff)-1):
        solution = [0] * (len(coeff)- 1)
        x = symbols(var[i])
        expr = coeff[i] * x + coeff[-1:][0]
        sol = solve(expr)
        solution[i] = float(sol[0]) if len(sol) > 0 else 0
        solution.append(coeff[-1:][0])
        reference_points.append(solution)
    return reference_points
```

```python
eqns = [[3.0, 4.0, 58.0, 500.0],
        [0.0, 5.0, 8.0, -550.0],
        [48.0, 30.0, -6.0, 120.0],
        [5.0, 6.0, -7.0, 25.0]]
```
